```python
# Keeping the descriptions in english only
df = df[df['description_lang'] == 'en']

# Resetting the index
df = df.reset_index()

# Dropping the new index column
df = df.drop('index', axis=1)

# Deleting all the rows with no descriptions 
df = df[df['description_narrative'].notnull()]

# We need to count the words in each description to keep only the longest 

# Creating a list with all the descriptions 
sentences = [i for i in df['description_narrative']]

# Setting an index list 
index = [i for i in range(len(sentences))] 

# Creating an empty list for the number of words which will later be fed 
number_words = []

#looping in the indexes
for i in index:
    #split the words in every sentence
    word = len(sentences[i].split())
    #appending the number of words in the list number_words
    number_words.append(word)

# Adding the list as a column to the main dataframe 
df['number_words'] = number_words

# View the dataset 
df.head()
```

```python
# See the main statistics of the descriptions 
df.describe()
```

```python
# We keep the longest descriptions
df = df[df['number_words'] > 207]
```

```python
# Randomly selecting 5,000 descriptions of the dataframe
df = df.sample(n = 5000)
```

```python
# create excel writer object
writer = pd.ExcelWriter('documents.xlsx')
# write dataframe to excel
df.to_excel(writer)
# save the excel
writer.save()
print('DataFrame is written successfully to Excel File.')
```
